    
 4) Factorial of a Number

    Problem:
    Given an integer n, calculate its factorial (n!) 

    Input: 5  
    Output: 120

```python
def fact(n):
    if n==0 or n==1:
        return 1
    else :
        return n * fact(n-1)

print(fact(5))
```

  6) Replace All Negative Numbers with Zero

    Given a list of integers, replace all negative numbers with 0

    Input: [-3, 5, -1, 7]  
    Output: [0, 5, 0, 7]

```python
 l= [-3, 5, -1, 7]  
 s=[]
 for n in l:
     if n < 0:
         s.append(0)
     else:
         s.append(n)
 s
```
