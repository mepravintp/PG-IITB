```python
def mean(l):
    sum=0
    for i in l:
        sum+=i
    return sum/len(l)
```

```python
print(mean(range(0,10,2)))
```

```python
def sdev(l):
    m=mean(l)
    result=0.0
    for x in l:
        result+=(x-m)** 2
    return np.sqrt(result/(len(l)-1))
```

```python
print(sdev(L))
```
