```python
# Q1 : Convolutional Neural Network (CNN)
# >>> data = np.arange(5)
# >>> kernel = [1/3, 1/3, 1/3]
# >>> print(convolution(data, kernel)))
# [0.33333333 1. 2. 3. 2.33333333]
# >>> print(convolution(data, kernel, stride=2))
# [0.33333333 2. 2.33333333]


#def convolution(data, kernel, stride):
#[0,1,2,3,4]
#

kernel = [1/3, 1/3, 1/3]
stride=1
# 0 1 2 = 0.33*0 + 1*0.33 + 2 *0.33
#1 2 3
# 2 3 4
# 



def convolution(data, kernel, stride=1):
    output_size=len(data)-len(kernel) +1
    output=[]
    for i in range(0, output_size,stride):
        
        slwindow=data[i:i+len(kernel)]
       # print(slwindow)
        sum=0
        for j in range(len(kernel)):
            
            sum=sum+ (slwindow[j]*kernel[j])
           # print(" J , Sum" , j ,sum)
        output.append(sum)
      
        
    return output        

convolution(np.arange(5),kernel,stride)
```

```python
def convolution1(data, kernel, stride=1):
    output_size = len(data) - len(kernel) + 1
    output = []
    for i in range(0, output_size, stride):
        window = data[i:i+len(kernel)]
        result = sum(window[j] * kernel[j] for j in range(len(kernel)))
        output.append(result)
    return np.array(output)

data = np.arange(5)
kernel = [1/3, 1/3, 1/3]
print(convolution1(data, kernel))
```

```python
def max_pool1(input, stride=2):
    output = []
    for i in range(0, len(input), stride):
        window = input[i:i+stride]
        output.append(max(window))
    return output

# Example usage:
print(max_pool1([0, 1, 2, 3, 4], stride=2))  # Output: [1, 3, 4]
print(max_pool1(list(range(6)), stride=3))
```

```python
#  Q3 
# We next implement layer normalization, write a function layer_norm(data) which re-
# turns normalized data by computing the mean d_mean and standard deviation, d_sigma
# of the data and finding result[i] = (data[i] - d_mean)/d_sigma. 
# Q1 - 1
# Q2 - 3
# Q3 - 2
# Q4 - 1
# Q5 - 0
# Q6 - 2

d1=[1,2,3,4]
def layer_norm(data):
    result=[]
    mean=np.mean(data)
    std=np.std(data)
    for i in d1:
        #print(i)
        result.append((i-mean)/std)
    return result


layer_norm(d1)

```
