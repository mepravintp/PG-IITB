```python
#imports

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python

#Load Data
df= pd.read_csv('../data/sine-segment-perturbed.csv')
df.head()

#Prepare Data
X = df['x'].values.reshape(-1, 1)
y = df['y']

#Linear Regression
model = LinearRegression()
model.fit(X, y)   
y_pred = model.predict(X)  

#Plotting Linear Regression

plt.figure(figsize=(10, 6))
plt.scatter(X, y, label='True values')
plt.plot(X, y_pred, color='blue', label='Linear Regression')
plt.legend()

print("RÂ² score for Linear Regression:", r2_score(y, y_pred))

#Polynomial Regression
degree = 6 # You can experiment with 3, 7, etc.
model = make_pipeline(PolynomialFeatures(degree), LinearRegression())
model.fit(X, y)

y_pred = model.predict(X)

plt.plot(X, y_pred, color='red', label=f'Polynomial degree {degree}')
plt.legend()
plt.show()


print("RÂ² score for Polynomial Regression:", r2_score(y, y_pred))

```

```python

#practice for different degrees
for degree in [4, 6.5, 8, 7,10 ,14,20]:
    model = make_pipeline(PolynomialFeatures(degree), LinearRegression())
    model.fit(X, y)
    y_pred = model.predict(X)
    r2 = r2_score(y, y_pred)
    print(f"Degree {degree} â RÂ²: {r2:.4f}")
```
