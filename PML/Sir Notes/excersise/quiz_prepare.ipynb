```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
d=[1,2,3,4,5]

print(np.median(d))

print(np.percentile)

data=[2,2,2,2,2,3,3,3,4,4,5]

def pmf(data):
    values, counts=np.unique(data,return_counts=True)
    #print(values,counts)
    return values,counts
```

```python
values,counts=pmf(data)
print(counts[values==2])  # Count of 3's in data


def cdf(data):
    values, counts=pmf(data)
    prob=counts/counts.sum()
    cum_prob=np.cumsum(prob)
    return values,cum_prob

print(cdf(data))
```
