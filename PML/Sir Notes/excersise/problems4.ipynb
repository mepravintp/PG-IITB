1. Consider the 7 different batsman data provided to you. Compute the mean,
   median, mode, standard deviation for each of these batsmen. Store the names
   of the batsmen in a list. Then store the mean/median/mode for each batsman
   in the same order.  You should have 5 lists with you at the end of this.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from numpy import mean, median


dhoni=np.loadtxt('../data/dhoni.txt') 
gambhir=np.loadtxt('../data/gambhir.txt') 
kapil=np.loadtxt('../data/kapil_dev.txt')
kohli=np.loadtxt('../data/kohli.txt')
sehwag=np.loadtxt('../data/sehwag.txt')
rohit=np.loadtxt('../data/rohit.txt')
sachin=np.loadtxt('../data/sachin.txt')

names=['dhoni','gambhir','kapil','kohli','sehwag','rohit','sachin']
means=[mean(dhoni),mean(gambhir),mean(kapil),mean(kohli),mean(sehwag),mean(rohit),mean(sachin)]
medians=[median(dhoni),median(gambhir),median(kapil),median(kohli),median(sehwag),median(rohit),median(sachin)]
modes=[pd.Series(dhoni).mode()[0],pd.Series(gambhir).mode()[0],pd.Series(kapil).mode()[0],pd.Series(kohli).mode()[0],pd.Series(sehwag).mode()[0],pd.Series(rohit).mode()[0],pd.Series(sachin).mode()[0]]

means,medians,modes;
```

2. Make a barchart comparing each of the batsmen. That is make 3 plots, one
   showing mean of all the batsmen (with suitable ticks), one plot for the
   median and one for the mode.

```python



plt.bar(names,means,label='means')
plt.bar(names,medians,label='medians')
plt.bar(names,modes,label='modes', color='red')
plt.legend()
plt.grid()
```

3. Look up the documentation for `plt.errorbar` (you can look this up online
   also) and draw the means of each batsman along with an error bar
   illustrating the standard deviation of their scores.

```python
stds = [np.std(dhoni), np.std(gambhir), np.std(kapil), np.std(kohli), np.std(sehwag), np.std(rohit), np.std(sachin)]

plt.errorbar(names, means, yerr=stds, fmt='o', capsize=5, color='blue', label='Mean  Std')
plt.ylabel('Runs')
plt.title('Mean scores with Standard Deviation')
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.boxplot([dhoni, gambhir, kapil, kohli, sehwag, rohit, sachin], tick_labels=names)
plt.ylabel('Runs')
plt.title('Boxplot Comparison of All Batsmen')
plt.grid(True)
plt.show()
```

```python



map ={"sachin":sachin,"kohli":kohli,"dhoni":dhoni,"sehwag":sehwag,"rohit":rohit,"kapil":kapil,"gambhir":gambhir,"All":[dhoni, gambhir, kapil, kohli, sehwag, rohit, sachin]}

@interact(batsman=[ i for i in map], plotType=["Histogram","Boxplot","Violin"],bins=['auto','fd','doane','scott','stone','rice','sturges','sqrt',5,10,15,20,25,30])
def showPlot(batsman,plotType,bins):
    print(batsman,plotType,bins)
    if batsman not in map:
        print("No data")
        return  
    
    data=map[batsman]
   # print(max(data))
    
    if plotType=="Histogram":
      plt.hist(data, bins=bins,label=batsman,density=True, edgecolor='black',align='left');
      plt.xlabel('Runs')
      plt.legend()
    elif plotType=="Boxplot":
        plt.boxplot(data, vert=True)
    elif plotType=="Violin":
        plt.violinplot(data, vert=True, showmeans=False, showmedians=True)
```
